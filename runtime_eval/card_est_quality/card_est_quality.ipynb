```python
import pandas 
df = pd.read_csv('/mnt/ml4qo/baseline/ssb-benchmark/results.csv')
est_quality = {}
for i in range(15):
    actual_rows = int(df['cardinality_true'].iloc[i])
    est_rows = int(df['cardinality_predict'].iloc[i] * 50)
    q_error = max(actual_rows / est_rows, est_rows / actual_rows)
    est_quality[f'{i+1}.sql'] = {
                'actual': actual_rows, 
                'estimated': est_rows,
                'estimation diff': '{0:+}'.format(est_rows - actual_rows),
                'q-error': q_error
                }
print(est_quality)
import json
with open('./eval_results/deepdb_ssb_sf10.json', 'w') as outfile:
    json.dump(est_quality, outfile, indent=4)

```
